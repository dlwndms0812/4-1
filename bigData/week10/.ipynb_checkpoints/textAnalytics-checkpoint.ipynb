```python
#필요한 파이썬 패키지 import

import pandas as pd
import glob
import re
from functools import reduce
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import STOPWORDS, WordCloud
```

```python
import nltk
nltk.download()
```

```python
#10개의 파일을 하나로 병합
all_files = glob.glob('data/myCabinetExcelData*.xls')
all_files
```

```python
all_files_data = []

for file in all_files :
    data_frame = pd.read_excel(file)
    all_files_data.append(data_frame)
all_files_data[0]
```

```python
all_files_data_concat = pd.concat(all_files_data, axis = 0, ignore_index = True)
all_files_data_concat
```

```python
all_files_data_concat.to_csv('data/riss_bigdata.csv', encoding = 'utf-8-sig', index = False)
#1~100까지의 인덱스만 반복됨
```

```python
#데이터 전처리
all_title = all_files_data_concat['제목']
all_title
```

```python
stopWords = set(stopwords.words("english")) #영어 불용어를 불러와서 저장
lemma = WordNetLemmatizer()
```

```python
words = []

for title in all_title :
    EnWords = re.sub(r"[^a-zA-Z]+"," ", str(title)) #알파벳 이외의 값은 공백으로 대체
    EnwordsToken = word_tokenize(EnWords.lower()) #위의 결과를 소문자로 바꾼 후, 토큰화
    EnWordsTokenStop = [w for w in EnwordsToken if w not in stopWords] #불용어 제거
    EnWordsTokenStopLemma = [lemma.lemmatize(w) for w in EnWordsTokenStop] #표제어 추출
    words.append(EnWordsTokenStopLemma)
```

```python
#2차원 리스트를 1차원 리스트로
words2 = list(reduce(lambda x, y : x + y, words))
print(words2)
```

```python
#단어 빈도 구하기
count = Counter(words2) #word2 리스트에 있는 단어 별로 빈도 수를 계산하여 결과로 딕셔너리 생성
count
```

```python
word_count = dict()

#출현 횟수가 많은 상위 50개 단어 추출
for tag, counts in count.most_common(50) :
    if (len(str(tag)) > 1) :
        word_count[tag] = counts
        print("{} : {}".format(tag, counts))
```

```python
#단어 빈도에 대한 탐색
del word_count['big']
del word_count['data']
plt.bar(range(len(word_count)), word_count.values(), align = 'center')
plt.xticks(range(len(word_count)), word_count.keys(), rotation = 'vertical')
plt.savefig('data/bar.jpg')
#plt.show()
```

```python
#연도별 'big data' 관련된 학술문서 수를 추출하여 그래프 그리기
all_files_data_concat['doc_count'] = 0
summary_year = all_files_data_concat.groupby('출판일', as_index = False)['doc_count'].count()
summary_year
```

```python
plt.figure(figsize = (12, 5))
plt.xlabel('year')
plt.ylabel('doc-count')
plt.grid(True)
plt.plot(range(len(summary_year)), summary_year['doc_count'])
plt.xticks(range(len(summary_year)), summary_year['출판일'])
plt.show()
```

```python
#워드 클라우드 그리기
stopwords = set(STOPWORDS)
wc = WordCloud(background_color = 'ivory', stopwords = stopwords, width = 800, height = 600)
cloud = wc.generate_from_frequencies(word_count)
plt.figure(figsize = (8, 8))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```

```python
cloud.to_file("data/riss_bigdata_wordCloud.jpg")
```
