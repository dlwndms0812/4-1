```python
#데이터 파일 확인
import pandas as pd
import numpy as np

data = pd.read_csv('공공보건 의료기관 현황_20161231.csv', index_col = 0, encoding = 'cp949', engine = 'python')
data.head()
```

```python
#병원명 분리하기

#연번 없이 병원 명만 따로 가져오기 힘들어 부득이 csv로 바꾼 후, '연번'과 '설립형태'를 제거하고 다시 값을 가져옴
tmp = data[['병원 명', '설립형태']]
tmp.to_csv('tmp.csv', encoding = 'cp949', index = True)
```

```python
#수동으로 가공한 csv를 다시 가져옴
hospital = pd.read_csv('tmp.csv', encoding = 'cp949', engine = 'python')

hospital
```

```python
#주소 정리하기
addr = pd.DataFrame(data['주소'].apply(lambda v : v.split(' ', 2)[0 : 3]).tolist(), columns = ('시도', '구군', '상세주소'))
addr.head()
```

```python
addr['시도'].unique()
```

```python
#주소 체계에 맞지 않게 표현된 내용 찾아서 수정

#'시'가 아닌 데이터 찾아서 수정
addr[addr['시도'] == '창원시']
```

```python
addr.iloc[27] = ['경상남도', '창원시', '의창구 북면 단내로 499-32']
addr.iloc[31] = ['경상남도', '창원시', '마산합포구3.15대로 231']
addr['시도'].unique()
```

```python
addr[addr['시도'] == '경산시']
```

```python
addr.iloc[47] = ['경상북도', '경산시', '경안로 23-10']
```

```python
addr[addr['시도'] == '천안시']
```

```python
addr.iloc[209] = ['충청남도', '천안시', '동남구 충절로 537(삼룡동)']
addr.iloc[210] = ['충청남도', '천안시', '동남구 목천읍 삼방로 735(천안시립노인병원)']
```

```python
addr['시도'].unique()
```

```python
#행정구역 주소 체계에 맞지 않는 데이터 수정
addr_aliases = {'경기' : '경기도', '경남' : '경상남도', '경북' : '경상북도', '충북' : '충청북도',
                '서울시' : '서울특별시', '부산특별시' : '부산광역시', '대전시' : '대전광역시',
                '충남' : '충청남도', '전남' : '전라남도', '전북' : '전라북도'}
addr['시도'] = addr['시도'].apply(lambda s : addr_aliases.get(s, s))

addr['시도'].unique()
```

```python
#구군 열 수정
addr['구군'].unique()
```

```python
addr[addr['구군'] == '아란13길']
```

```python
addr.iloc[75] = ['제주특별자치도', '제주시', '아란13길 15(아라일동)']
```

```python
#모든 주소 합치기
addr['총 주소'] = addr.apply(lambda r : r['시도'] + ' ' + r['구군'] + ' ' + r['상세주소'], axis = 1)
addr.head()
```

```python
#병원명 더하기
addr_merge = pd.merge(addr, hospital, left_index = True, right_index = True)
addr_merge.head()
```

```python
#새로운 csv 만들기
addr_merge.to_csv('Address.csv', encoding = 'cp949', index = False)
```

```python
#지도에 표시하기
import folium

cbGeoData = pd.read_csv('Address_new.csv', encoding = 'utf-8', engine = 'python')
mapCB = folium.Map(location = [37.5555038, 126.9737511], zoom_start = 15)

for i, arr in cbGeoData.iterrows() :
    #_Y는 위도 _X는 경도
    folium.Marker(location = [arr['_Y'], arr['_X']], 
                  popup = arr['field5'], icon = folium.Icon(color = 'red', icon = 'star')).add_to(mapCB)
mapCB.save('./hostipal_map.html')
```
